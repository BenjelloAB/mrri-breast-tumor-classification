```python
import requests
import time

def get_series_metadata_by_uid(series_uid):
    url = "https://services.cancerimagingarchive.net/services/v4/TCIA/query/getSeries"
    params = {"SeriesInstanceUID": series_uid, "format": "json"}
    response = requests.get(url, params=params)
    if response.status_code == 200 and response.json():
        return response.json()[0]  # TCIA returns a list
    return None
get_series_metadata_by_uid(series_uids[0])
```

```python
import requests
import zipfile
import os

def download_series(series_uid, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    api_url = f"https://services.cancerimagingarchive.net/services/v4/TCIA/query/getImage?SeriesInstanceUID={series_uid}"
    try:
        response = requests.get(api_url, stream=True, timeout=30)
        if response.status_code != 200:
            print(f"Failed series {series_uid}: HTTP {response.status_code}")
            return 0
        zip_path = os.path.join(output_dir, f"{series_uid}.zip")
        with open(zip_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024):
                f.write(chunk)
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(output_dir)
        os.remove(zip_path)
        dicom_files = [f for f in os.listdir(output_dir) if f.endswith(".dcm")]
        print(f"Downloaded {len(dicom_files)} DICOMs for series {series_uid}")
        return len(dicom_files)
    except Exception as e:
        print(f"Error downloading series {series_uid}: {str(e)}")
        return 0
```

```python
from google.colab import drive
import shutil
import time


series_uids = [meta["SeriesInstanceUID"] for meta in mri_series_metadata]

# Output directory
output_base_dir = "/content/bd_dataset"
os.makedirs(output_base_dir, exist_ok=True)

# Download in batches
batch_size = 10  # Process 10 series at a time
total_dicom_count = 0

for batch_start in range(0, len(series_uids), batch_size):
    batch_uids = series_uids[batch_start:batch_start + batch_size]
    temp_dir = "/content/temp_series"
    os.makedirs(temp_dir, exist_ok=True)

    print(f"Processing batch {batch_start//batch_size + 1}/{(len(series_uids) + batch_size - 1)//batch_size}")
    for i, series_uid in enumerate(batch_uids):
        dicom_count = download_series(series_uid, os.path.join(temp_dir, series_uid))
        total_dicom_count += dicom_count
        print(f"Series {batch_start + i + 1}/{len(series_uids)} complete")
        time.sleep(1)  # Avoid API rate limits

    # Move to folder
    for series_uid in batch_uids:
        src = os.path.join(temp_dir, series_uid)
        dst = os.path.join(output_base_dir, series_uid)
        if os.path.exists(src):
            shutil.move(src, dst)

    # Clear temporary directory
    shutil.rmtree(temp_dir, ignore_errors=True)

print(f"Total DICOMs downloaded: {total_dicom_count}")
```

```python

df = df_transposed
def map_pathology_to_class(label):
    if pd.isna(label):
        return 'Unknown'

    label = label.lower()

    benign_keywords = ['benign', 'fibroadenoma', 'fibrosis', 'hyperplasia', 'fibrocyst']
    malignant_keywords = ['invasive', 'infiltrating', 'intraductal', 'carcinoma', 'ca',
                          'ductal', 'lobular', 'in-situ', 'grade', 'scarff-bloom-richardson', 'component']

    if any(b in label for b in benign_keywords) and not any(m in label for m in malignant_keywords):
        return 'Benign'
    elif any(m in label for m in malignant_keywords):
        return 'Malignant'
    else:
        return 'Unknown'
df['Diagnosis'] = df['Pathology'].apply(map_pathology_to_class)



df['Diagnosis'].unique()
print(df['Diagnosis'].value_counts())
print(f"Patients : {df['PatientID'].unique()}")
print(f"Number of Patients {len(df['PatientID'].unique())}")
df
```

```python
import pydicom
import os

def check_dicom_modality(base_dir="./bd_dataset"):
    non_mr_files = []
    for series_dir in os.listdir(base_dir):
        series_path = os.path.join(base_dir, series_dir)
        if os.path.isdir(series_path):
            for dicom_file in os.listdir(series_path):
                if dicom_file.endswith('.dcm'):
                    dicom_path = os.path.join(series_path, dicom_file)
                    try:
                        ds = pydicom.dcmread(dicom_path)
                        modality = ds.Modality
                        if modality != 'MR':
                            non_mr_files.append((dicom_path, modality))
                    except Exception as e:
                        print(f"Error reading {dicom_path}: {str(e)}")
    if non_mr_files:
        print(f"Found {len(non_mr_files)} non-MR DICOMs:")
        for path, modality in non_mr_files:
            print(f"{path}: Modality = {modality}")
    else:
        print("All DICOMs are MR (MRI).")
    return non_mr_files

non_mr_files = check_dicom_modality()
```

```python
# Slice 1: Balance + Encode Labels + Split
# Goal: Balance to 12,012 Benign, 12,012 Malignant (24,024), split into train (16,826), val (3,594), test (3,604).

import os
import shutil
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from PIL import Image
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras import layers, models, callbacks
# Encode labels: Benign=0, Malignant=1
df_final["label_encoded"] = df_final["label"].map({"Benign": 0, "Malignant": 1})

# Balance
df_mal = df_final[df_final['label'] == 'Malignant']
df_ben = df_final[df_final['label'] == 'Benign']
df_mal_down = df_mal.sample(n=len(df_ben), random_state=42)
df_balanced = pd.concat([df_ben, df_mal_down]).sample(frac=1, random_state=42).reset_index(drop=True)

# Split: 70% train, 15% val, 15% test
df_temp, df_test = train_test_split(df_balanced, test_size=0.15, stratify=df_balanced["label_encoded"], random_state=42)
df_train, df_val = train_test_split(df_temp, test_size=0.176, stratify=df_temp["label_encoded"], random_state=42)

# Verify
print(df_balanced['label'].value_counts())
print(f"Train: {len(df_train)}, Val: {len(df_val)}, Test: {len(df_test)}")

# Save splits
df_train.to_csv('train_split.csv', index=False)
df_val.to_csv('val_split.csv', index=False)
df_test.to_csv('test_split.csv', index=False)

```

```python
# Slice 2: DICOM to JPEG + Delete DICOM
# Goal: Convert ~24,024 DICOMs to JPEGs (224x224, RGB, quality=60), delete DICOMs, store ~3–6GB.

import pandas as pd
import pydicom
import numpy as np
from PIL import Image
import os
import shutil
import multiprocessing as mp
from functools import partial

def convert_dicom_to_jpeg(row, output_dir, size=(224, 224)):
    try:
        dcm_path = row['image_path']
        dcm = pydicom.dcmread(dcm_path)
        img_array = dcm.pixel_array
        img_array = (img_array - img_array.min()) / (img_array.max() - img_array.min() + 1e-7) * 255
        img_array = img_array.astype(np.uint8)
        img = Image.fromarray(img_array).convert("RGB").resize(size)
        label = row['label']
        out_path = os.path.join(output_dir, label, f"{row['PatientID']}_{row.name}.jpg")
        img.save(out_path, quality=60)
        os.remove(dcm_path)
        return row.name, out_path
    except Exception as e:
        print(f"Failed on {row['image_path']}: {e}")
        return row.name, None

def preprocess_split(df, split_name, output_dir, num_workers=2):
    split_dir = os.path.join(output_dir, split_name)
    for label in ['Benign', 'Malignant']:
        os.makedirs(os.path.join(split_dir, label), exist_ok=True)

    pool = mp.Pool(num_workers)
    convert_func = partial(convert_dicom_to_jpeg, output_dir=split_dir)
    results = pool.map(convert_func, [row for _, row in df.iterrows()])
    pool.close()
    pool.join()

    for idx, out_path in results:
        if out_path:
            df.at[idx, 'image_path'] = out_path
    df = df.dropna(subset=['image_path'])
    return df

# Setup output directory
output_dir = "/content/skin_images_224"
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir)

# Load splits
df_train = pd.read_csv('train_split.csv')
df_val = pd.read_csv('val_split.csv')
df_test = pd.read_csv('test_split.csv')

# Preprocess
df_train = preprocess_split(df_train, 'train', output_dir)
df_val = preprocess_split(df_val, 'val', output_dir)
df_test = preprocess_split(df_test, 'test', output_dir)

# Save updated CSVs
df_train.to_csv('train_split_processed.csv', index=False)
df_val.to_csv('val_split_processed.csv', index=False)
df_test.to_csv('test_split_processed.csv', index=False)

# Verify disk
print(f"Disk usage: {os.popen('df -h .').read()}")

```
